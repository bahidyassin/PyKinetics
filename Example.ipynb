# Basic Setup


```python
KS1 = pk.KineticSystem(1) # number of subsystems

KS1.AddInteraction(pk.TransitionRight(0.09,0.5),1,1) 
```

```python
## Interpolate
KS1.Interpolate(100)
```

```python
## Initial Value
mean1  = 0.5
std1 = 0.01

def init1(x):
    return np.exp(-(x-mean1)**2/std1 )
```

```python
KS1.Solve([init1], limit_time= 20)


KS1.MultiPlot(1,[0,5,10, 20]) # ADD name = 'path' to save 
```

## MAssive System

```python
KSN = pk.KineticSystem(7)

KSN.AddInteraction(pk.TransitionLeftXY(0.9,0.5),1,2,1) 

KSN.AddInteraction(pk.TransitionChi2ChiXY(0,0.1, 0.8, 1,10),1,2,2) 

KSN.AddInteraction(pk.TransitionLeftXY(0.4,0.5),2,1,2) 

KSN.AddInteraction(pk.TransitionLeftXY(0.4,0.5),3,4,6) 

KSN.AddInteraction(pk.TransitionRightXY(0.4,0.5),5,4,5) 

KSN.AddInteraction(pk.TransitionRight(0.4,0.5),6,5)

KSN.AddInteraction(pk.TransitionRightXY(0.4,0.5),5,4,5) 

KSN.AddInteraction(pk.TransitionAwayXY(0.4,0.5,0.6),5,4,5)

KSN.AddInteraction(pk.TransitionLeftXY(0.4,0.5),6,7,6)

KSN.AddInteraction(pk.TransitionRight(0.4,0.5),7,7)
```

```python
st_time = time.time()

KSN.Interpolate(100)

mean1  = 0.5
std1 = 0.01
mean2 = 0.5
std2 = 0.01
mean3  = 0.5
std3 = 0.01
mean4 = 0.2
std4 = 0.01
mean5 = 0.8
std5 = 0.01
mean6 = 0.7
std6 = 0.01
mean7 = 0.2
std7 = 0.03
def init1(x):
    # print(x)
    return np.exp(-(x-mean1)**2/std1 )
def init2(x):
    return np.exp(-(x-mean2)**2/std2)
def init3(x):
    return np.exp(-(x-mean3)**2/std3)
def init4(x):
    return np.exp(-(x-mean4)**2/std4)  + np.exp(-(x-0.8)**2/std4)
def init5(x):
    return np.exp(-(x-mean5)**2/std5)
def init6(x):
    return np.exp(-(x-mean6)**2/std6)
def init7(x):
    return np.exp(-(x-mean7)**2/std7)

KSN.Solve([init1, init2, init3, init4, init5, init6, init7], limit_time= 100)

print(time.time() - st_time)
```

# O2O Model

```python
KSN = pk.KineticSystem(3)


### Subsystem 1:
KSN.AddInteraction(pk.TransitionAwayXY(0.4,0.5,0.6),1,2,1)
KSN.AddInteraction(pk.TransitionAwayXY(0.4,0.5,0.6),1,3,1)
KSN.AddInteraction(pk.TransitionChi2ChiXY(0.9,1, 0, 0.1,10),1,2,2) 

### Subsystem 2 :

KSN.AddInteraction(pk.TransitionRightXY(0.7,1),2,2,2)
KSN.AddInteraction(pk.TransitionRightXY(0.7,1),2,3,2)

KSN.AddInteraction(pk.TransitionChi2ChiXY(0.9,1, 0.9, 1,20),2,2,3) 

###Subsystem 3 :


KSN.AddInteraction(pk.TransitionLeft(0.09,2),3,3)
KSN.AddInteraction(pk.TransitionChi2ChiXY(0,.2, 0.9, 1,10),3,3,1) 
```

```python
KSN.Interpolate(100)

mean1  = 0.5
std1 = 0.01
mean2 = 0.5
std2 = 0.01
mean3  = 0.8
std3 = 0.001

def init1(x):
    # print(x)
    return np.exp(-(x-mean1)**2/std1 )
def init2(x):
    return np.exp(-(x-mean2)**2/std2)
def init3(x):
    return 0.5*np.exp(-(x-mean3)**2/std3)
```

```python
KSN.Solve([init1, init2, init3], limit_time=10)
```

```python
KSN.MultiPlot(3,[0,5,10])
```

```python
KSN.Animate(1, 'GP', duration = 10, fps = 12, TimeLabel = True)
KSN.Animate(2, 'Online', duration = 10, fps = 12, TimeLabel = True)
KSN.Animate(3, 'Offline', duration = 10, fps = 12, TimeLabel = True)
```
